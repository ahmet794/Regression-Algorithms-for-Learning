```python
from sklearn.linear_model import Lasso
import gradient_descent as gd
import data_utils
from data_utils import StandardScaler

%load_ext autoreload
%autoreload 2
```

```python
train_data, test_data = data_utils.load_boston(42)

scaler = StandardScaler()
train_normalized = scaler.fit_transform(train_data)
test_normalized = scaler.transform(test_data)
```

```python
import mglearn
lasso = Lasso(alpha=0.001)

X, y = mglearn.datasets.load_extended_boston()

lasso.fit(X, y)

lasso.coef_
lasso.score(X, y)
```

```python
gd.lasso_gradient_descent(train_normalized, 0.0001, 0.01, 100, 0.001)
```

```python
from ridge_regression import RidgeRegression
import matplotlib.pyplot as plt

rr = RidgeRegression(0.001)

model = rr.train(train_normalized)

rr_coefs = model['Beta Hat']

plt.figure(figsize=(6,4))
plt.bar(range(len(rr_coefs)), rr_coefs, color = "green")
plt.xlabel("Coefficient Index")
plt.ylabel("Coefficient Value")
plt.title(f"Ridge Regression Coefficients $\lambda$ = {rr.alpha}")
plt.show()

model["R-Squared Training"]

```
