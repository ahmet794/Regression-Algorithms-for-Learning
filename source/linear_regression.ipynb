```python
np.random.seed(70)  # for reproducibility

# Independent variable: Quality score (ranging from 1 to 10)
quality_score = np.linspace(1, 10, 100)

# Dependent variable: Price
# We'll create a linear relationship with some noise added
price = 5 * quality_score + 3 + np.random.normal(0, 2, 100)

# Convert to DataFrame for better visualization
df = pd.DataFrame({
    'Quality_Score': quality_score,
    'Price': price
})

data = df.to_numpy()
data
```

```python
np.random.shuffle(data)

train_size = int(data.shape[0] * 0.8)
train_data = data[:train_size]
test_data = data[train_size:]

print(train_data)
test_data
```

```python
lr.slr(train_data, test_data)
```

Comparing the mlr and slr should give similar intercept and slope and R-Squared.

```python
print(lr.train_mlr(train_data))
lr.slr(train_data, test_data)
```

We can see that we have similar answers for mlr and slr as required.

```python
beta_hat = lr.train_mlr(train_data)["Beta_hat"]
lr.predict_mlr(test_data, beta_hat)
```
